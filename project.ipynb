```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('image.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
img_blur = cv2.GaussianBlur(image_gray,(3,3),0 )
edges = cv2.Canny(image = img_blur ,threshold1=100,threshold2=200)
plt.imshow(image)
plt.imshow(edges)
```

```python
import cv2
import matplotlib.pyplot as plt
import cvlib as cv
import numpy as np
im = cv2.imread('couple.jpeg')
plt.imshow(im)
faces, conf = cv.detect_face(img) 
for face in faces:
    (startX,startY) = face[0],face[1]
    (endX,endY) = face[2],face[3]
    cv2.rectangle(img, (startX,startY), (endX,endY), (0,255,0), 2)
                  

plt.imshow(img)
plt.show()

```

```python
import cv2
import matplotlib.pyplot as plt

import numpy as np
img = cv2.imread('blood smear.png')
gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite('gray.png', gray)
blur = cv2.medianBlur(gray,5)
cv2.imwrite('blur.png',blur)
blurg = cv2.GaussianBlur(gray,(9,9), 0)
cv2.imwrite('blurg.png',blurg)
histo = cv2.equalizeHist(gray)
cv2.imwrite('histo norm.png', histo)
clahe = cv2.createCLAHE(clipLimit = 2.0 , tileGridSize = (8,8))
claheNorm = clahe.apply(gray)
cv2.imwrite('claheNorm.png',claheNorm)


```
